# Retail Sales Analysis: The Impact of Promotions
- In this project, we study the relationship between promotions and sales in a retail dataset.
- The objective is to understand whether promotions are associated with higher sales for a given product.
- We use a simplified approach by focusing on a subset of the data.

```python
import pandas as pd

# We load only a subset of the dataset to keep things simple
df = pd.read_csv("C:/Users/rapha/OneDrive/Documents/PSE x ENPC/Programming and Data Analytics/train.csv")

# Display first rows
df.head()
```

Let's display the first rows of the dataset to understand its structure.
-> Each row corresponds to the sales of a specific product in a given store at a given date.
--> For example, the 1st row shows that on January 1st, 2013, store 25 sold 7 units of product 103665.

The dataset contains several important variables:

- `date`: the date of the observation  
- `store_nbr`: the store identifier  
- `item_nbr`: the product identifier  
- `unit_sales`: the number of units sold  
- `onpromotion`: indicates whether the product was on promotion  

We can see that on a given date, a store sells multiple different products, each represented by a different row.

This confirms that the dataset is detailed at the product-store-day level.
